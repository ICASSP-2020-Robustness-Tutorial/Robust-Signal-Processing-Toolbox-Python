# Examples for Bounded Innovation Propagation (BIP) MM ARMA parameter estimation

```python
import numpy as np
import scipy.io as sio
import scipy.signal as sps
import robustsp as rsp
import matplotlib.pyplot as plt
import matplotlib

# Fix random number generator for reproducibility
np.random.seed(1) 
```

## Example 1: AR(1) with 30 percent isolated outliers

```python
# Generate AR(1) observations
N = 300
a = np.random.randn(N)
x = sps.lfilter([1],[1,.8],a)
p = 1
q = 0
```

### Generate isolated Outliers

```python
cont_prob = 0.3 # outlier contamination probability
outlier_ind =  np.where(np.sign(np.random.randn(N)-cont_prob)<0)# outlier index
outlier = 100*np.random.randn(N) # contaminating process
v = np.zeros(N) # additive outlier signal
v[outlier_ind] = outlier[outlier_ind]
v[0] = 0 # first sample should not be an outlier

x_ao = x+v # 30% of isolated additive outliers

x = sio.loadmat('x')['x'].flatten()
x_ao = sio.loadmat('x_ao')['x_ao'].flatten() # max(x) = 4.6821
```

### BIP MM Estimation

```python
result = rsp.arma_est_bip_mm(x_ao,p,q) # result['cleaned_signal'] falsch
```

```python
print('Example: AR(1) with ar_coeff = 0.8')
print('30% of isolated additive outliers')
print('estimaed coefficients: %.3f' % result['ar_coeffs'])

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [10, 10]

plt.subplot(2,1,1)
plt.plot(x_ao,'-',lw=2,c='r',label='outlier contaminated AR(1)')
plt.plot(result['cleaned_signal'],'-.',c='y',label='cleaned')

plt.xlabel('samples')
plt.ylabel('Amplitude')
plt.title('BIP-AR(1) cleaned signal')
plt.legend()

plt.subplot(2,1,2)
plt.plot(x,lw=2,label='original AR(1)')
plt.plot(result['cleaned_signal'],'-.',c='y',label='cleaned')

plt.xlabel('samples')
plt.ylabel('Amplitude')
plt.title('BIP-AR(1) cleaned signal')
plt.legend()
plt.show()
```

# Example 2: ARMA(1,1) with 10% patchy outliers

## Generate ARMA(1,1) observations

```python
N = 1000
a = np.random.randn(N)
x = sps.lfilter([1, 0.2],[1, -.8],a)
p = 1
q = 1
```

## Generate a patch of outliers of length 101 samples

```python
v = 1000*np.random.randn(101)
```

## 10% of patch additive outliers

```python
x.shape
```

```python
x_ao = np.array(x)
x_ao[99:199] += v[99:199]
```

```python
a,b,c= rsp.bip_s_resid_sc(x_ao,[-.75],1,0)
res1 = rsp.arma_est_bip_s(x_ao,p,q)
np.max(b),np.max(x),np.max(result['cleaned_signal']),np.max(res1['cleaned_signal'])
```

```python
# Robust starting point by BIP AR-S approximation
from scipy.optimize import least_squares as lsq
beta_initial = rsp.robust_starting_point(x_ao,p,q)[0] 

F_bip = lambda beta: rsp.bip_s_resid_sc(x_ao, beta, p, q)[2]
beta_bip = lsq(F_bip, beta_initial,xtol=5*1e-7,ftol=5*1e-7,method='lm')['x']

a_bip_sc, x_filt2,_ = rsp.bip_s_resid_sc(x_ao, beta_bip, p, q)

beta_bip, np.max(x_filt2)
```

```python
x_filt = rsp.ar_est_bip_s(x_ao,p)[1] # slightly wrong
import statsmodels.tsa.api as tsa
# TODO make estimates like matlab
mod = tsa.SARIMAX(x_filt, order=(p, 0, q), concentrate_scale=True,
                  enforce_stationarity=False, enforce_invertibility=False)
res = mod.fit()
beta_initial= res.params
beta_initial
```
