```python
import robustsp as rsp
import numpy as np

X =  np.asarray\
    ([[2,     3],
      [4,     6],
      [9,     8]])

def signcm(Xx,center=False,EPS=1e-6,):
    X = np.array(Xx)
    
    if center: 
        smed0 = rsp.spatmed(X)
        X = X - smed0[:,None]
    else: smed0 = []
        
    len = np.sqrt(np.sum(X*np.conj(X),axis=1))
    X[len != 0,:] = X
    len[len!=0] = len
    n,p = X.shape
    len[len<EPS]=EPS
    X = X / len[:,None]
    return X.T @ np.conj(X)/n , smed0

signcm(X)
```
